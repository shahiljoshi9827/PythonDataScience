```python
#functional

plt.plot(x,y,'b--')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('XY REPRESENTATION')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g--')
plt.show()
```

```python
plt.hist(y,x,orientation='vertical')
```

```python
# object oriented

fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes1=fig.add_axes([0.25,0.2,0.5,0.5])
axes.plot(x,y)
axes1.plot(y,x)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
#for current_ax in axes:
#    current_ax.plot(x,y)
plt.plot(x,y)
plt.plot(y,x)
plt.tight_layout()
```

```python
plt.plot(y,x)
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('1st')
axes[1].plot(y,x)
axes[1].set_title('2nd')

plt.tight_layout()
```

```python
#figuresize and dpi
```

```python
fig=plt.figure(figsize=(5,5),dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,'r--')
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,alpha=0.5,ls='-',marker='o',markersize=10,markerfacecolor='y',markeredgewidth=3,markeredgecolor='k')
ax.plot(y,x,color='b',linewidth=3,alpha=0.5,ls='-',marker='o',markersize=10,markerfacecolor='y',markeredgewidth=3,markeredgecolor='k')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
plt.show()
```
